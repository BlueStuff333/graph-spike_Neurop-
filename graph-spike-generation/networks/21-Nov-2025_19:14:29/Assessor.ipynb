```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
```

```python
data = loadmat("graph-spike_.mat")
print(data.keys())
```

```python
firings = data['firings']
print(type(firings))
print(firings.shape)

plt.plot(firings[:, 0], firings[:, 1], 'k|', markersize=2)
plt.xlabel('Time')
plt.ylabel('Neuron Index')
plt.show()
```
