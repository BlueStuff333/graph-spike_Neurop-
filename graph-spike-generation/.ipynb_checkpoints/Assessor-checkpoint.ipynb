```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = loadmat("graph-spike_.mat")
print(data.keys())
```

```python
color_data = loadmat("graph_node_colors.mat")
node_colors = color_data['node_colors']
```

```python
firings = data['firings']
print(type(firings))
print(firings.shape)

plt.figure(figsize=(10, 6))
for neuron_idx in np.unique(firings[:, 1].astype(int)):
    mask = firings[:, 1] == neuron_idx
    color = node_colors[neuron_idx - 1]
    plt.plot(firings[mask, 0], firings[mask, 1], '|', 
             color=color, markersize=2)
plt.xlabel('Time')
plt.ylabel('Neuron Index')
plt.title('Raster Plot (Color-coded by Community)')
plt.show()
```

```python
unique_colors = np.unique(node_colors, axis=0)
color_to_cluster = {tuple(c): i for i, c in enumerate(unique_colors)}
neuron_to_cluster = np.array([color_to_cluster[tuple(node_colors[i])] 
                               for i in range(len(node_colors))])

sorted_neurons = np.argsort(neuron_to_cluster)
neuron_map = {old_idx: new_idx for new_idx, old_idx in enumerate(sorted_neurons)}

plt.figure(figsize=(10, 6))
for spike in firings:
    neuron_idx = int(spike[1]) - 1
    new_idx = neuron_map[neuron_idx]
    color = node_colors[neuron_idx]
    plt.plot(spike[0], new_idx, '|', color=color, markersize=2)

plt.xlabel('Time')
plt.ylabel('Neuron Index (Clustered by Community)')
plt.title('Raster Plot (Neurons Clustered by Community)')
plt.show()
```
